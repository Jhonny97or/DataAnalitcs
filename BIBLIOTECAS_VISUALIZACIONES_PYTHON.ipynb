# 3 BIBLIOTECAS DE PYTHON MAS USADAS PARA CREAR VISUALIZACIONES

Maplotlib (10 visualizaciones)

**1.1 Gráfico de dispersión:**

Descripción: Muestra la relación entre dos variables numéricas.
Uso: Ideal para identificar patrones, correlaciones y valores atípicos en los datos.
Ejemplo: Analizar la relación entre la edad y el salario de los empleados de una empresa.

**1.2 Gráfico de barras:**

Descripción: Muestra la comparación de valores entre diferentes categorías.
Uso: Ideal para representar datos categóricos y realizar comparaciones entre grupos.
Ejemplo: Comparar las ventas mensuales de diferentes productos en una tienda.

**1.3 Gráfico de violín:**

Descripción: Representa la distribución de datos y su densidad.
Uso: Ideal para visualizar la distribución de una variable numérica en diferentes categorías.
Ejemplo: Analizar la distribución de los puntajes de estudiantes en diferentes asignaturas.

**1.4 Gráfico de líneas:**

Descripción: Muestra la evolución de una variable a lo largo del tiempo.
Uso: Ideal para visualizar tendencias y cambios en valores numéricos a lo largo de un período.
Ejemplo: Seguir el crecimiento de la población en una ciudad durante los últimos 20 años.

**1.5 Gráfico de área:**

Descripción: Muestra la distribución acumulativa de una variable a lo largo del tiempo.
Uso: Ideal para resaltar cambios en la composición de un conjunto de datos a lo largo del tiempo.
Ejemplo: Visualizar la proporción de ingresos por sector económico en un país a lo largo de décadas.

**1.6 Gráfico de tarta:**

Descripción: Representa la proporción de cada categoría en un conjunto de datos.
Uso: Ideal para mostrar la composición relativa de diferentes partes en un todo.
Ejemplo: Mostrar la distribución por género de los empleados de una empresa.

**1.7 Gráfico de burbujas:**

Descripción: Muestra la relación entre tres variables utilizando el tamaño y el color de las burbujas.
Uso: Ideal para visualizar patrones y relaciones complejas entre múltiples variables.
Ejemplo: Analizar la relación entre el PIB per cápita, la esperanza de vida y la población de diferentes países.

**1.8 Gráfico de mapas de calor:**

Descripción: Muestra la distribución de datos en una matriz utilizando colores.
Uso: Ideal para identificar patrones y correlaciones en grandes conjuntos de datos.
Ejemplo: Analizar la correlación entre diferentes variables en un estudio de investigación.

**1.9 Gráfico de treemap:**

Descripción: Muestra la jerarquía y proporciones de datos en una estructura de árbol.
Uso: Ideal para visualizar la distribución proporcional de categorías y subcategorías.
Ejemplo: Representar la estructura organizativa de una empresa con diferentes departamentos y equipos.

**1.10 Gráfico de radar:**

Descripción: Muestra la comparación de múltiples variables utilizando ejes radiales.
Uso: Ideal para comparar el rendimiento o características de múltiples elementos.
Ejemplo: Evaluar las habilidades en diferentes áreas de un grupo de candidatos en un proceso de selección.



```python
import numpy as np
import matplotlib.pyplot as plt

# Ejemplo 1: Gráfico de líneas
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='Función Seno')
plt.title("Gráfico de Líneas - Función Seno")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.grid(True)
plt.show()

# Ejemplo 2: Gráfico de dispersión
x = np.random.rand(100)
y = np.random.rand(100)
plt.scatter(x, y, color='red', marker='o', label='Puntos Aleatorios')
plt.title("Gráfico de Dispersión - Puntos Aleatorios")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.grid(True)
plt.show()

# Ejemplo 3: Gráfico de barras
x = ["A", "B", "C", "D", "E"]
y = [10, 15, 7, 12, 9]
plt.bar(x, y, color='green')
plt.title("Gráfico de Barras - Ventas por Categoría")
plt.xlabel("Categorías")
plt.ylabel("Ventas")
plt.grid(True)
plt.show()

# Ejemplo 4: Histograma
data = np.random.randn(1000)
plt.hist(data, bins=30, color='orange', alpha=0.7)
plt.title("Histograma - Distribución de Puntajes")
plt.xlabel("Puntajes")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()

# Ejemplo 5: Gráfico de pastel
labels = ["A", "B", "C", "D", "E"]
sizes = [15, 30, 20, 10, 25]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'pink', 'yellow', 'lightcoral'])
plt.title("Gráfico de Pastel - Distribución de Ventas")
plt.show()

# Ejemplo 6: Gráfico de área
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.fill_between(x, y1, y2, alpha=0.5, color='purple', label='Área entre Funciones')
plt.title("Gráfico de Área - Funciones Seno y Coseno")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.grid(True)
plt.show()

# Ejemplo 7: Gráfico de caja y bigotes
data = np.random.randn(100)
plt.boxplot(data, vert=False, notch=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Gráfico de Caja y Bigotes - Distribución de Puntajes")
plt.xlabel("Puntajes")
plt.grid(True)
plt.show()

# Ejemplo 8: Gráfico de violin
data = np.random.randn(100)
plt.violinplot(data, vert=False, widths=0.5, showmedians=True, showextrema=True)
plt.title("Gráfico de Violin - Distribución de Puntajes")
plt.xlabel("Puntajes")
plt.grid(True)
plt.show()

# Ejemplo 9: Gráfico de contorno
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
plt.contourf(X, Y, Z, cmap='cool')
plt.title("Gráfico de Contorno - Superficie Sinusoidal")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.colorbar()
plt.grid(True)
plt.show()

# Ejemplo 10: Gráfico de heatmap
data = np.random.rand(10, 10)
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.title("Gráfico de Heatmap - Datos Aleatorios")
plt.colorbar(label="Valores")
plt.grid(True)
plt.show()
```

**Seaborn (10 visualizaciones)**

**2.1 Gráfico de distribución:**

Descripción: Visualiza la distribución de datos univariados o bivariados.
Uso: Ideal para analizar la forma y la dispersión de los datos.
Ejemplo: Observar la distribución de las calificaciones de un examen en una clase.

**2.2 Gráfico de conteo:**

Descripción: Muestra el recuento de observaciones en diferentes categorías.
Uso: Ideal para visualizar la frecuencia de categorías en un conjunto de datos.
Ejemplo: Verificar la cantidad de compras realizadas en diferentes días de la semana.

**2.3 Gráfico de caja y bigotes:**

Descripción: Representa la distribución de datos y los valores atípicos.
Uso: Ideal para analizar la dispersión y la presencia de valores atípicos en diferentes grupos o categorías.
Ejemplo: Comparar la distribución de salarios en diferentes departamentos de una empresa.

**2.4 Gráfico de violín:**

Descripción: Combina un gráfico de caja y bigotes con un gráfico de densidad.
Uso: Ideal para visualizar la distribución y la concentración de valores en diferentes categorías.
Ejemplo: Comparar la distribución de las calificaciones de estudiantes en diferentes asignaturas.

**2.5 Gráfico de regresión:**

Descripción: Muestra la relación entre dos variables con una línea de regresión.
Uso: Ideal para visualizar la tendencia y la fuerza de la relación entre variables continuas.
Ejemplo: Analizar la relación entre la edad y el salario de los empleados de una empresa y ajustar una línea de regresión.

**2.6 Heatmap:**

Descripción: Representa la matriz de datos utilizando colores para resaltar valores.
Uso: Ideal para visualizar la relación y la estructura de los datos en una matriz.
Ejemplo: Analizar la correlación entre diferentes variables en un conjunto de datos.

**2.7 Gráfico de correlación:**

Descripción: Muestra la matriz de correlación entre variables numéricas.
Uso: Ideal para identificar relaciones lineales o no lineales entre variables.
Ejemplo: Evaluar la correlación entre el índice de obesidad y el consumo de alimentos en diferentes países.

**2.8 Pairplot:**

Descripción: Crea una matriz de gráficos de dispersión para todas las combinaciones de variables.
Uso: Ideal para explorar la relación entre múltiples variables de forma simultánea.
Ejemplo: Analizar la relación entre la edad, el salario y la educación en una muestra de empleados.

**2.9 Gráfico de barras apiladas:**

Descripción: Muestra la comparación de valores apilados de diferentes categorías.
Uso: Ideal para representar la contribución relativa de subcategorías dentro de una categoría principal.
Ejemplo: Comparar los ingresos por categoría de productos en diferentes regiones.

**2.10 Gráfico de línea de tiempo:**

Descripción: Visualiza la evolución de una variable a lo largo del tiempo.
Uso: Ideal para mostrar tendencias y patrones temporales.
Ejemplo: Seguimiento del crecimiento de suscriptores en una plataforma de streaming a lo largo de varios años.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Ejemplo 1: Gráfico de dispersión con línea de regresión
tips = sns.load_dataset("tips")
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title("Gráfico de Dispersión con Línea de Regresión - Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()

# Ejemplo 2: Gráfico de barras
titanic = sns.load_dataset("titanic")
sns.countplot(x="class", data=titanic)
plt.title("Gráfico de Barras - Pasajeros por Clase")
plt.xlabel("Clase")
plt.ylabel("Cantidad de Pasajeros")
plt.show()

# Ejemplo 3: Gráfico de violín
iris = sns.load_dataset("iris")
sns.violinplot(x="species", y="sepal_length", data=iris)
plt.title("Gráfico de Violín - Longitud del Sépalo por Especie")
plt.xlabel("Especie")
plt.ylabel("Longitud del Sépalo")
plt.show()

# Ejemplo 4: Gráfico de caja y bigotes
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Gráfico de Caja y Bigotes - Total Bill por Día")
plt.xlabel("Día")
plt.ylabel("Total Bill")
plt.show()

# Ejemplo 5: Gráfico de conteo por categoría
titanic = sns.load_dataset("titanic")
sns.catplot(x="class", hue="survived", kind="count", data=titanic)
plt.title("Gráfico de Conteo por Categoría - Supervivientes por Clase")
plt.xlabel("Clase")
plt.ylabel("Cantidad de Pasajeros")
plt.show()

# Ejemplo 6: Gráfico de densidad
iris = sns.load_dataset("iris")
sns.kdeplot(data=iris, x="sepal_width", y="sepal_length", cmap="Blues")
plt.title("Gráfico de Densidad - Longitud y Ancho del Sépalo")
plt.xlabel("Ancho del Sépalo")
plt.ylabel("Longitud del Sépalo")
plt.show()

# Ejemplo 7: Gráfico de correlación (heatmap)
flights = sns.load_dataset("flights").pivot("month", "year", "passengers")
sns.heatmap(flights, cmap="YlOrRd", annot=True, fmt="d")
plt.title("Gráfico de Correlación - Pasajeros por Mes y Año")
plt.xlabel("Año")
plt.ylabel("Mes")
plt.show()

# Ejemplo 8: Gráfico de línea con intervalo de confianza
fmri = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", hue="region", data=fmri)
plt.title("Gráfico de Línea con Intervalo de Confianza - Señal en el Tiempo")
plt.xlabel("Tiempo")
plt.ylabel("Señal")
plt.show()

# Ejemplo 9: Gráfico de jointplot
iris = sns.load_dataset("iris")
sns.jointplot(data=iris, x="sepal_length", y="sepal_width", kind="hex")
plt.title("Gráfico de Jointplot - Longitud y Ancho del Sépalo")
plt.xlabel("Longitud del Sépalo")
plt.ylabel("Ancho del Sépalo")
plt.show()

# Ejemplo 10: Gráfico de pairplot
iris = sns.load_dataset("iris")
sns.pairplot(data=iris, hue="species")
plt.title("Gráfico de Pairplot - Relación entre Variables")
plt.show()
```

**Plotly (10 visualizaciones)**

**3.1 Gráfico de dispersión 3D:**

Descripción: Representa la relación entre tres variables en un espacio tridimensional.
Uso: Ideal para visualizar patrones y relaciones complejas en datos tridimensionales.
Ejemplo: Analizar la relación entre la temperatura, la humedad y la presión atmosférica en diferentes ubicaciones geográficas.

**3.2 Gráfico de superficie 3D:**

Descripción: Muestra una superficie tridimensional a partir de datos en una matriz.
Uso: Ideal para visualizar datos que varían en dos dimensiones espaciales y una dimensión de valor.
Ejemplo: Representar la topografía de un terreno utilizando datos de altitud.

**3.3 Gráfico de contorno 2D:**

Descripción: Muestra las líneas de contorno de datos en un plano bidimensional.
Uso: Ideal para visualizar la distribución y los cambios de valor en un espacio bidimensional.
Ejemplo: Visualizar las zonas de temperatura alta y baja en un mapa meteorológico.

**3.4 Gráfico de mapa de calor:**

Descripción: Representa los valores de una matriz con colores en un plano bidimensional.
Uso: Ideal para visualizar la distribución y la intensidad de los valores en una matriz.
Ejemplo: Mostrar la concentración de población en diferentes regiones geográficas.

**3.5 Gráfico de área apilada:**

Descripción: Muestra la contribución relativa de diferentes categorías a través del tiempo con áreas apiladas.
Uso: Ideal para comparar la evolución de múltiples categorías a lo largo del tiempo.
Ejemplo: Visualizar los ingresos generados por diferentes productos en un período de tiempo determinado.

**3.6 Gráfico de treemap:**

Descripción: Representa jerarquías de datos utilizando áreas rectangulares anidadas.
Uso: Ideal para mostrar la estructura y la relación entre categorías jerárquicas.
Ejemplo: Visualizar la estructura de gastos en un presupuesto familiar, donde cada nivel jerárquico representa una categoría de gasto.

**3.7 Gráfico de radar:**

Descripción: Muestra múltiples variables en un gráfico circular con ejes radiales.
Uso: Ideal para comparar el desempeño o características de diferentes entidades en varias categorías.
Ejemplo: Comparar las habilidades de diferentes jugadores de fútbol en diferentes aspectos del juego.

**3.8 Gráfico de embudo:**

Descripción: Representa la progresión y la tasa de conversión en un proceso secuencial.
Uso: Ideal para visualizar la eficiencia y la pérdida en etapas sucesivas de un proceso.
Ejemplo: Analizar el flujo de clientes en un embudo de ventas y determinar la tasa de conversión en cada etapa.

**3.9 Gráfico de histograma 2D:**

Descripción: Muestra la distribución conjunta de dos variables continuas utilizando colores.
Uso: Ideal para visualizar la relación entre dos variables continuas y su densidad de distribución.
Ejemplo: Analizar la relación entre la edad y el nivel de ingresos en una muestra de población.

**3.10 Gráfico de violín:**

Descripción: Combina un gráfico de caja con una representación de la densidad de los datos.
Uso: Ideal para visualizar la distribución de datos y comparar múltiples categorías.
Ejemplo: Comparar la distribución de las calificaciones de diferentes grupos de estudiantes en un examen.

```python
import plotly.express as px
import plotly.graph_objects as go

# Ejemplo 1: Gráfico de dispersión
df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", title="Gráfico de Dispersión")
fig.show()

# Ejemplo 2: Gráfico de barras
df = px.data.tips()
fig = px.bar(df, x="day", y="total_bill", title="Gráfico de Barras")
fig.show()

# Ejemplo 3: Gráfico de violín
df = px.data.tips()
fig = px.violin(df, y="tip", x="day", box=True, title="Gráfico de Violín")
fig.show()

# Ejemplo 4: Gráfico de líneas
df = px.data.gapminder().query("continent == 'Asia'")
fig = px.line(df, x="year", y="lifeExp", color="country", title="Gráfico de Líneas")
fig.show()

# Ejemplo 5: Gráfico de área
df = px.data.gapminder().query("continent == 'Asia'")
fig = px.area(df, x="year", y="pop", color="country", title="Gráfico de Área")
fig.show()

# Ejemplo 6: Gráfico de tarta
df = px.data.tips()
fig = px.pie(df, values="tip", names="day", title="Gráfico de Tarta")
fig.show()

# Ejemplo 7: Gráfico de burbujas
df = px.data.gapminder()
fig = px.scatter(df, x="gdpPercap", y="lifeExp", size="pop", color="continent", title="Gráfico de Burbujas")
fig.show()

# Ejemplo 8: Gráfico de mapas de calor
df = px.data.tips()
fig = px.density_heatmap(df, x="total_bill", y="tip", title="Gráfico de Mapas de Calor")
fig.show()

# Ejemplo 9: Gráfico de treemap
df = px.data.tips()
fig = px.treemap(df, path=['day', 'time', 'sex'], values='total_bill', title="Gráfico de Treemap")
fig.show()

# Ejemplo 10: Gráfico de radar
categories = ['sales', 'marketing', 'development', 'customer support', 'IT', 'administration']
values = [400, 350, 300, 450, 200, 100]
fig = go.Figure(data=go.Scatterpolar(r=values, theta=categories, fill='toself'))
fig.update_layout(title="Gráfico de Radar")
fig.show()
```
